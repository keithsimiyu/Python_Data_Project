```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\KEITH\Downloads\data_jobs.csv")
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df["job_skills"].apply(lambda x:ast.literal_eval(x) if pd.notna(x)else x)
```

```python
df_DA = df[df["job_title_short"]=="Data Analyst"].copy()
```

```python
df_DA
```

```python
df_DA["job_posted_no"] = df_DA["job_posted_date"].dt.month
df_DA
```

```python

df_DA_explode = df_DA.explode("job_skills")

df_DA_pivot = df_DA_explode.pivot_table(index="job_posted_no", columns="job_skills",aggfunc='size',fill_value =0)
```

```python
df_DA_pivot.head(10)
```

```python
# since we want the top skills
# lets create a row for the total, find the sum and sort 
df_DA_pivot.loc["Total"] = df_DA_pivot.sum()
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc["Total"].sort_values(ascending=False).index]
df_DA_pivot = df_DA_pivot.drop("Total")
df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind="line")
plt.title("Top 5 skills for data analyst per month")
plt.ylabel("count")
plt.xlabel(' ')
plt.show()
```
