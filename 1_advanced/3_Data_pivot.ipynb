```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\KEITH\Downloads\data_jobs.csv")
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# working with pivot tables
df.pivot_table(index="job_title_short", aggfunc="size")

##you can well say use group by but it will have some limitations late
##ie df.groupby("job_title_short").size()
```

```python
df.pivot_table(values="salary_year_avg",index="job_title_short", aggfunc="median")
##you as well say df.groupby("job_title_short")["salary_year_avg"].median()
```

```python
# now finding the median salary of different job titles broken down by country
df.pivot_table(values="salary_year_avg",index="job_country",columns='job_title_short', aggfunc="median")
##interchanged the index and columns so that job country are on index and job tit on columns
```

```python
##finding top six countries
top_countries= df["job_country"].value_counts().head(6).index
top_countries
```

```python
df_job_country_salary =df.pivot_table(values="salary_year_avg",index="job_country",columns='job_title_short', aggfunc="median")
df_job_country_salary = df_job_country_salary.loc[top_countries]
df_job_country_salary
```

```python
job_titles = ["Data Analyst","Data Scientist", "Data Engineer"]
df_job_country_salary = df_job_country_salary[job_titles]
df_job_country_salary
```

```python
import matplotlib.pyplot as plt
```

```python
##we can run a plot on it easily now

df_job_country_salary.plot(kind="bar")
plt.ylabel("Median Salary($USD)")
plt.xlabel(' ')
plt.title("Median salary by country job")
plt.xticks(rotation=45,ha="right")
plt.show()
```
